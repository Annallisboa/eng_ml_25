```python
rows = (
    df
    [
    ['lat','lon','minutes_remaining', 'period', 'playoffs', 'shot_distance','shot_made_flag']
    ].dropna()
).drop(columns=["shot_made_flag"]).values.tolist()
```

```python
import requests

data = {
    "dataframe_split": {
        "columns": ["lat", "lon", "minutes_remaining", "period", "playoffs", "shot_distance"],
        "data": rows
    }
}
response = requests.post(
    "http://localhost:5001/invocations",
    json=data,
    headers={"Content-Type": "application/json"}
)
predicoes = response.json()
predicoes["predictions"][:2]
```
